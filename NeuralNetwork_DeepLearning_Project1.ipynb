Drop the columns which are unique for all users like IDs (2.5 points)

```python
bank_df.drop(columns = ['Surname'],inplace=True)
```

```python
#Finding the number of counts in each categorical columns
for i in bank_df.columns:
    if bank_df[i].dtype == 'O':
        print(i,'->',len(bank_df[i].value_counts()))
```

```python
#Manual Encoding
bank_df.replace({'France':0,'Germany':1,'Spain':2,'Male':0,'Female':1},inplace=True)
#bank_df.replace({'France':0,'Germany':1,'Spain':2},inplace=True)
```

Distinguish the feature and target set (2.5 points)

```python
from sklearn.model_selection import train_test_split

X = bank_df.drop("Exited", axis=1)
X = X.drop({'Geography','Gender'}, axis=1)
X = X.astype('float32')
```

```python
y = bank_df[['Exited']]
y = y.astype('float32')
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X.dtype
```

Divide the data set into Train and test sets

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

Normalize the train and test data (2.5 points)

```python
%tensorflow_version 2.x
```

```python
import tensorflow as tf
tf.random.set_seed(42)
```

```python
tf.__version__
```

```python
import keras
```

Initialize & build the model (7.5 points)

```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer()
train_set = transformer.fit_transform(train_set)
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

Optimize the model (5 points)

```python
model.fit(train_set, train_labels, 
          validation_data=(test_set, test_labels), 
          epochs=100,
          batch_size=32)
```

```python
model.summary()
```

Predict the results using 0.5 as a threshold (5 points)

```python
w = tf.zeros(shape=(8,1))
b = tf.zeros(shape=(1))
```

```python
def prediction(x, w, b):
    
    xw_matmul = tf.matmul(x, w)
    y = tf.add(xw_matmul, b)
    
    return y
```

```python
def loss(y_actual, y_predicted):
    
    diff = y_actual - y_predicted
    sqr = tf.square(diff)
    avg = tf.reduce_mean(sqr)
    
    return avg
```

```python
def train(x, y_actual, w, b, learning_rate=0.5):
    
    #Record mathematical operations on 'tape' to calculate loss
    with tf.GradientTape() as t:
        
        t.watch([w,b])
        
        current_prediction = prediction(x, w, b)
        current_loss = loss(y_actual, current_prediction)
    
    #Calculate Gradients for Loss with respect to Weights and Bias
    dw, db = t.gradient(current_loss,[w, b])
    
    #Update Weights and Bias
    w = w - learning_rate*dw
    b = b - learning_rate*db
    
    return w, b
```

```python
for i in range(100):
    
    w, b = train(train_set, train_labels, w, b,learning_rate=0.5)
    print('Current Loss on iteration', i, loss(train_labels, prediction(train_set, w, b)).numpy())
```

Print the Accuracy score and confusion matrix (5 points)

```python
from sklearn.metrics import accuracy_score
y_predict = model.predict_classes(test_set, verbose=0)
```

```python
# accuracy: (tp + tn) / (p + n)
accuracy = accuracy_score(test_labels, y_predict)
print('Accuracy: %f' % accuracy)
```

```python
# precision tp / (tp + fp)
from sklearn.metrics import precision_score
precision = precision_score(test_labels, y_predict)
print('Precision: %f' % precision)

```

```python
# recall: tp / (tp + fn)
from sklearn.metrics import recall_score
recall = recall_score(test_labels, y_predict)
print('Recall: %f' % recall)
```

```python
# f1: 2 tp / (2 tp + fp + fn)
from sklearn.metrics import f1_score
f1 = f1_score(test_labels, y_predict)
print('F1 score: %f' % f1)
```

```python
from sklearn.metrics import confusion_matrix
# confusion matrix
matrix = confusion_matrix(test_labels, y_predict)
print(matrix)
```

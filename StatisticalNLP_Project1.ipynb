```python
import pandas as pd       
blog = pd.read_csv("blogtext.csv",nrows=5000)
```

```python
blog.head(5)
```

```python
blog.drop(columns=['id', 'date'],inplace=True)
```

```python
blog = blog.apply(lambda x: x.astype(str).str.lower())
```

```python
import re
blog = blog.applymap(lambda s: re.sub("[^a-zA-Z0-9#+_]"," ",s))
```

```python
blog["text"] = blog["text"].str.strip()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
# set of stop words
stop = set(stopwords.words('english')) 
```

```python
blog["text"] = blog["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
blog.head(5)
```

```python
blog['labels']= blog["gender"].astype(str) +","+ blog["age"] +","+ blog["topic"] +","+ blog["sign"]
```

```python
blog.drop(columns =['gender', 'age', 'topic', 'sign'],inplace=True)
```

```python
blog.head(2)
```

```python
blog.shape
```

```python
blog.info()
```

```python
blog['labels'] = blog['labels'].astype('category')
```

```python
blog['text'] = blog['text'].astype('category')
```

```python
X = blog['text']
```

```python
y = blog['labels']
```

```python
from sklearn.model_selection import train_test_split # Splitting the data for training and testing out model

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state = 1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1, 2),stop_words='english',min_df=2)
#min_df
```

```python
train_vectors = vectorizer.fit_transform(X_train)
#vectorizer.fit_transform isa sparse matrix.It means it will only store the non zero element of a matrix.
test_vectors = vectorizer.transform(X_test)
```

```python
print(train_vectors.shape, test_vectors.shape)
```

```python
print(vectorizer)
```

```python
print(train_vectors)
```

```python
count_label=vectorizer.vocabulary_
print(count_label)
```

```python
blog.labels.values
```

```python
vectorizer_labels = CountVectorizer(min_df = 2,ngram_range = (1,1),stop_words = "english")
```

```python
labels_vector = vectorizer_labels.fit_transform(y)
```

```python
vectorizer_labels.vocabulary_
```

```python
# Extracing only key value from above dictionary, which contains unique labels. These set of labels will be used as classes in 
# multilabelbinariser further.
label_classes = []  
for key in vectorizer_labels.vocabulary_.keys():
    label_classes.append(key)
    
print(label_classes)   
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer(classes = label_classes)  
```

```python
##Convert Y_train into a format as required by mlb 
y_train = [["".join(re.findall("\w",f)) for f in lst] for lst in [s.split(",") for s in y_train]]
y_train[0]
```

```python
y_train_mlb = mlb.fit_transform(y_train)
y_train_mlb
```

```python
y_train_mlb.dtype
```

```python
mlb.classes_
```

```python
'''changed mlb.fit_transform to mlb.transform.When you call fit() or fit_transform(), the mlb forgets its previous learnt 
data and learn the new supplied data. This can be problematic when Y_train and Y_test may have difference in labels as my 
case have.'''
Y_test = [["".join(re.findall("\w",f)) for f in lst] for lst in [s.split(",") for s in y_test]]
y_test_mlb = mlb.transform(Y_test)
y_test_mlb
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
clf = LogisticRegression(solver='lbfgs',max_iter=400)
clf = OneVsRestClassifier(clf)
```

```python
clf.fit(train_vectors, y_train_mlb)
```

```python
from sklearn.metrics import accuracy_score
score_train = clf.score(train_vectors, y_train_mlb)
print(score_train)
```

```python
from sklearn.metrics import accuracy_score
score_test = clf.score(test_vectors, y_test_mlb)
print(score_test)
```

```python
predicted = clf.predict(test_vectors)
print(str(accuracy_score(y_test_mlb, predicted)))
```

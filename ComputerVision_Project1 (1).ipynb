```python
x_train = []
y_train = []
```

```python
#read and resize the image
import cv2
path=os.getcwd()
for i in os.listdir(path):
    print(i)
    if (os.path.isdir(i)):
            for j in os.listdir(i):
                try:
                    dummy = cv2.imread('/content/drive/My Drive/Colab Notebooks/ComputerVision_Project1/train/' + i + "/" + j)
                    dummy = cv2.resize(dummy,(128,128))
                    x_train.append(dummy)
                    y_train.append(i)
                except Exception as e:
                    print(e)
```

```python
x_train[0].shape
```

```python
import pandas as pd
dum = pd.get_dummies(y_train)
```

```python
encoded_labels = dum
y_train = dum
```

```python
import numpy as np
y_train = np.array(y_train)
```

```python
x_train = np.array(x_train)
```

```python
x_train[0].shape
```

```python
#Divide the data set into Train and validation data sets
from sklearn.model_selection import train_test_split
x_train2, x_val, y_train2, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=2)
print (len(x_train2))
print (len(x_val))
```

```python
x_train2[0].shape
```

```python
x_train2 = x_train2.reshape(x_train2.shape[0],128,128,3)
```

```python
x_val = x_val.reshape(x_val.shape[0],128,128,3)
```

```python
x_train2 = x_train2/255.

x_val = x_val/255.
```

```python
print (x_train2.shape)
print (x_val.shape)
```

```python
print(y_train2.shape)
print(y_val.shape)
```

```python
from keras import applications
from keras.models import Sequential, Model 
from keras.layers import Dropout, Flatten, Dense
from keras import backend as k 
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers import BatchNormalization
```

```python
#Initialize & build the model 
scale = 128
seed = 7
model = Sequential()
model.add(Conv2D(filters=128, kernel_size=(5, 5), input_shape=(scale, scale, 3), activation='relu'))
model.add(BatchNormalization(axis=3))
model.add(Conv2D(filters=128, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(BatchNormalization(axis=3))
model.add(Dropout(0.1))

model.add(Conv2D(filters=256, kernel_size=(5, 5), activation='relu'))
model.add(BatchNormalization(axis=3))
model.add(Conv2D(filters=256, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(BatchNormalization(axis=3))
model.add(Dropout(0.1))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

classes = 12
model.add(Dense(classes, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

Optimize the model

```python
# Train the model 
model.fit(x_train2, y_train2, epochs = 1, validation_data=(x_val, y_val))
```

```python
#Predict the accuracy for validation data
loss_and_metrics = model.evaluate(x_val, y_val)
print(loss_and_metrics)
```

```python
#Predict the accuracy for both train data
loss_and_metrics_train = model.evaluate(x_train2, y_train2)
print(loss_and_metrics_train)
```
